Installing libraries

```python
import pandas as pd
import altair as alt
from vega_datasets import data as vega_data
```

Loading Seattle Circulation data

```python
import requests
import io

url = "https://raw.githubusercontent.com/manika-lamba/SP26-LIS4_5693/refs/heads/main/lab-assignments/lab-3/Seattle-Library_2015-2021.csv"
response = requests.get(url)
response.raise_for_status() # Raise an exception for HTTP errors
text = response.text
```

Saving as dataframe

```python
seattle_df = pd.read_csv(io.StringIO(text))
seattle_df.head()
```

## Exercise

Task One and Two: Visualizing 2 Different Columns with Differing Altair Marks

```python
seattle_df
```

Visualizing Usage Class counts over the years

```python
seattle_grouped_by_usage = seattle_df.groupby(['CheckoutYear', 'UsageClass']).size().reset_index(name='counts')

alt.Chart(seattle_grouped_by_usage).mark_line(point=True).encode(
    x=alt.X('CheckoutYear:O', title='Checkout Year'),
    y=alt.Y('counts:Q', title='Count'),
    color=alt.Color('UsageClass:N', title='Usage Class'),
).properties(
    title='Physical vs Digital Checkouts Over the Years'
)
```

Visualizing Checkouts by Month per Publisher

```python
alt.data_transformers.disable_max_rows()

alt.Chart(seattle_df).mark_circle(size=80).encode(
    x=alt.X('CheckoutMonth:O', title='Checkout Month'),
    y=alt.Y('count():Q', title='Number of Checkout Records'),
    color=alt.Color('Publisher:N', title='Publisher'),
)
```

Task Three Adding Interactivity

Adding zooming to Visualization 1

```python
alt.Chart(seattle_grouped_by_usage).mark_line(point=True).encode(
    x=alt.X('CheckoutYear:O', title='Checkout Year'),
    y=alt.Y('counts:Q', title='Count'),
    color=alt.Color('UsageClass:N', title='Usage Class'),
).properties(
    title='Physical vs Digital Checkouts Over the Years'
).interactive()
```

Adding tooltips and interactivity to Visualization 2

```python
input_dropdown = alt.binding_select(
    options=sorted(seattle_df['Publisher'].dropna().unique()),
    name='Publisher '
)

selection = alt.selection_point(
    fields=['Publisher'],
    bind=input_dropdown
)

color = (
    alt.when(selection)
    .then(alt.Color('Publisher:N').legend(None))
    .otherwise(alt.value('lightgray'))
)
alt.Chart(seattle_df).mark_circle(size=80).encode(
    x=alt.X('CheckoutMonth:O', title='Checkout Month'),
    y=alt.Y('count():Q', title='Number of Checkout Records'),
    color=color,
    tooltip=['Publisher', 'CheckoutMonth', 'count()']
).add_params(
    selection
).properties(
    title='Count of Checkouts by Month for Each Publisher'
)
```

Task Four: Horizontal Concatenation

Naming Charts

```python
viz1 = alt.Chart(seattle_grouped_by_usage).mark_line(point=True).encode(
    x=alt.X('CheckoutYear:O', title='Checkout Year'),
    y=alt.Y('counts:Q', title='Count'),
    color=alt.Color('UsageClass:N', title='Usage Class'),
).properties(
    title='Physical vs Digital Checkouts Over the Years'
)
viz2 = alt.Chart(seattle_df).mark_circle(size=80).encode(
    x=alt.X('CheckoutMonth:O', title='Checkout Month'),
    y=alt.Y('count():Q', title='Number of Checkout Records'),
    color=alt.Color('Publisher:N', title='Publisher'),
)
```

Combining

```python
alt.hconcat(viz1,viz2)
```
